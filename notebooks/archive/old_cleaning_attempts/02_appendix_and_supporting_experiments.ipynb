## Observable-volatility Heston robustness check

This is a supporting robustness check, not the main Heston result. Direct access
to the simulated variance \(v_t\) is replaced by an EWMA proxy \(\widehat v_t\).
The neural network still observes the contemporaneous liquid-option quote
\(C^h_t/S_t\). That quote is observable at the hedging date, so this is not
future-information leakage, but it carries implied information about the latent
variance state. The experiment should therefore be read as an observable-market-
information robustness check, not as a pure stock-return-only volatility-filtering
experiment.

Purpose:
Measure how much of the Heston stock-and-option result survives when the hedge conditions on a causal EWMA variance proxy instead of true simulated variance.

Report use:
Appendix robustness check for the main full-information Heston result.

Key caveat:
This must not overwrite full-information Heston files.

Expected outputs:
Mode-safe observable-volatility outputs under `results/appendix/heston_obsvol/` and `figures/appendix/heston_obsvol/`.
